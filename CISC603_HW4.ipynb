```python
import pandas as pd

#NFA inputs
nfa = {}
n = int(input("Enter total number of states: ")) #input total number of states
p = int(input("Enter total number of paths: ")) #input total number of paths
for i in range(n):
    state = input("state name:")
    nfa[state] = {} #create NFA dictionary
    for j in range(p):
        path = input("path: ") #create path
        print("Enter the end state from state {} travels through path {}:".format(state,path))
        endState = [x for x in input().split()] #assign end state for each path
        nfa[state][path] = endState

        
#Printing NFA
print("\nNFA: \n")
print(nfa)
print("\nNFA table: ") #print NFA in table
nfa_table = pd.DataFrame(nfa)
print(nfa_table.transpose())



#DFA outputs
newStates = [] #stores all new states created in DFA
dfa = {} #create DFA dictionary
keysList = list(list(nfa.keys())[0]) #list of all states in NFA and DFA
pathList = list(nfa[keysList[0]].keys()) #list of all paths


#first row of DFA
dfa[keysList[0]] = {} #create DFA dictionary
for k in range(p):
    var = "".join(nfa[keysList[0]][pathList[k]]) #create new state which contains all elements 
    dfa[keysList[0]][pathList[k]] = var #assign the new state to DFA
    if var not in keysList: #if the state is new
        newStates.append(var) #append it to new_states_list
        keysList.append(var) #append to state list
        
    
while len(newStates) != 0: 
    dfa[newStates[0]] = {}
    for m in range(len(newStates[0])):
        for i in range(len(pathList)):
            temp = []
            for j in range(len(newStates[0])):
                temp += nfa[newStates[0][j]][pathList[i]]
            s = ""
            s = s.join(temp)
            if s not in keysList:
                newStates.append(s)
                keysList.append(s)
            dfa[newStates[0]][pathList[i]] = s
            
    newStates.remove(newStates[0])
                
        
print("\nDFA: \n")
print(dfa)
print("\nDFA table: ")
dfa_table = pd.DataFrame(dfa)
print(dfa_table.transpose())

    
                
                
```
